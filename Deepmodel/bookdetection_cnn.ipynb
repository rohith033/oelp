```python
import cv2
%matplotlib inline
import matplotlib.pyplot as plt
img1 = cv2.imread("./book11.jpeg")
img2 = cv2.imread("./book2.jpeg")
img1 = cv2.resize(img1,(600,50))
img2 = cv2.resize(img2,(600,50))
img=[]
img.append(img1)
img.append(img2)
fig , axis = plt.subplots(1,2,figsize=(16,16))
for i,ax in enumerate(axis.flat):
  ax.imshow(img[i])
```

```python
# make a white patch of 25*50
import numpy as np
tmp = np.ones((50,50,3),dtype=float)
print(tmp)
tmp = cv2.resize(tmp,(50,50))
white=[]
white.append(tmp)
white.append(tmp)
fig , axis = plt.subplots(1,2,figsize=(16,16))
for i,ax in enumerate(axis.flat):
  ax.imshow(white[1])
```

```python
crops=[]
for i in range(0,12):
  crop = img2[0:50,0+50*i:50+50*i]
  crops.append(crop)
fig,axis=plt.subplots(4,3,figsize=(16,16))
print(len(crops))
for i,ax in enumerate(axis.flat):
  ax.imshow(crops[i])
```

```python
def mse(img1, img2):
   h, w , z = img1.shape
   diff = cv2.subtract(img1, img2,dtype=0)
   err = np.sum(diff**2)
   mse = err/(float(h*w*z))
   return mse
error = 1e6
idx=-1
black = np.zeros((50,50,3))
image_to_extract=black
for i in range(len(crops)):
  chk = np.array(crops[i],np.float64)
  print(mse(chk,tmp))
  print(i)
  if(mse(chk,tmp)<error):
    # print(i)
    error=mse(crops[i],tmp)
    imgage_to_extract=crops[i]
    idx=i
print(idx)
```

```python
fig,axis=plt.subplots(1,2,figsize=(16,16))
for i,ax in enumerate(axis.flat):
  ax.imshow(imgage_to_extract)
```
